## 

- So, It recommends movies based on similarity. similarity of tags column.
- for example, below two movies tags are there. 
- how we can predict similarity between these two tags.
- What are the most common things or what common text between those two. 
- So, how can we calculate the similarity score between them.
- here comes, vectorization.

####

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=55000,stop_words='english')
```

```python
vector = cv.fit_transform(df['Tags']).toarray()
```

```python
vector
```

```python
vector[0]
```

```python
stop_words = pd.DataFrame(cv.get_feature_names_out())
```

```python
stop_words.values

for i in stop_words.values:
    print(i)
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index = df[df['Movie'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    for i in movie_list:
        print(df.iloc[i[0]].Movie)
```

```python
movies_names = df["Movie"]
movie_list = movies_names.tolist()  
```

```python
for i in range(0, len(movie_list), 4):
    print(movie_list[i:i+4])
```

```python
recommend("The Last: Naruto the Movie") 
```

```python
recommend("Spider-Man") 
```

```python
recommend("Re-Animator") 
```
