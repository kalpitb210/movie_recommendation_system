```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
titles = []
years = []
Genre = []
time = []
imdb_ratings = []
metascores = []
votes = []
movie_categories = []
us_gross = []
Director = []
Discription = []
Actor1 = []
Actor2 = []
Actor3 = []
Actor4 = []
images = []
```

```python
for i in range(1, 10159, 50):
    url = f'https://www.imdb.com/search/title/?title_type=feature&num_votes=10000,&sort=user_rating,desc&start={i}&ref_=adv_nxt'

    headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
               "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}

    r = requests.get(url, headers=headers).text
    soup = BeautifulSoup(r , "lxml")

    movie_div = soup.find_all('div', class_='lister-item mode-advanced')
    

    for container in movie_div:
        # Scraping the movie's name
        name = container.h3.a.text
        if name is None:
            titles.append(np.nan)
        else:
            titles.append(name)



        # Scraping the movie's Genre
        genre1 = container.find("span",  {"class":  "genre"}).getText().replace(" ", "").replace("\n", "")
        Genre.append(genre1)



        # Scraping the movie's year
        year = container.h3.find('span', class_='lister-item-year').text.replace("(","").replace(")","")
        if year is None:
            years.append(np.nan)
        else:
            years.append(year)




        # Scraping the movie's length
        runtime = container.find('span', class_='runtime').text if container.p.find('span', class_='runtime') else np.nan
        time.append(runtime)



        # Scraping the rating
        rating = container.find('div', class_='ratings-bar').strong.text
        if rating is None:
            imdb_ratings.append(np.nan)
        else:
            imdb_ratings.append(rating)



        # Scraping the movie's Director
        director = container.find("p", class_="").a.text.replace("\n",'').replace("Director:","")
        if director is None:
            Director.append(np.nan)
        else:
            Director.append(director)



        # Scraping the metascore
        m_score = container.find('span', class_='metascore').text if container.find('span', class_='metascore') else np.nan
        metascores.append(m_score)



        # Scraping votes and gross earnings
        nv = container.find_all('span', attrs={'name':'nv'})
        vote = nv[0].text
        votes.append(vote)
        grosses = nv[1].text if len(nv) > 1 else np.nan
        us_gross.append(grosses)



        # Scraping the movie's Discription
        discription = container.findAll("p", class_="text-muted")[1].text.replace("\n", "")
        if discription is None:
            Discription.append(np.nan)
        else:
            Discription.append(discription)



        # Scraping the movie's Actor1
        actor1 = container.find_all("p", class_="")[0].select_one("a:nth-of-type(2)")
        if actor1 is None:
            Actor1.append(np.nan)
        else:
            Actor1.append(actor1.text)



        # Scraping the movie's Actor2
        actor2 = container.find_all("p", class_="")[0].select_one("a:nth-of-type(3)")
        if actor2 is None:
            Actor2.append(np.nan)
        else:
            Actor2.append(actor2.text)



        # Scraping the movie's Actor3
        actor3 = container.find_all("p", class_="")[0].select_one("a:nth-of-type(4)")
        if actor3 == None:
            Actor3.append(np.nan)
        else:
            Actor3.append(actor3.text)



        # Scraping the movie's Actor4    
        actor4 = container.find_all("p", class_="")[0].select_one("a:nth-of-type(5)")
        actor4
        if actor4 is None:
            Actor4.append(np.nan)
        else:
            Actor4.append(actor4.text)




        # Scraping Movie_Poster 
        try:
            image_url = container.find('div', class_='lister-item-image float-left').a.img['src']
        except:
            image_url = np.nan
        images.append(image_url)

```

```python
temp_df = {"Movie": titles, "Year": years, "Rating": imdb_ratings,"Metascore": metascores, "Genre": Genre, 
           "Director": Director, "Actor1": Actor1, "Actor2": Actor2, "Actor3": Actor3, "Actor4": Actor4, 
           "Runtime": time, "Votes": votes, "Gross": us_gross, "Discription": Discription, "Poster": images}
```
